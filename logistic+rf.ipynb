```python
train = pd.read_csv("../input/train.csv")
test = pd.read_csv("../input/test.csv")
```

```python
train_sub = train.iloc[0:2000,]

train_sub.shape
train_sub.columns
```

```python
id = train_sub['ID_code']
y = train_sub['target']
x = train_sub.drop(['ID_code','target'],axis=1)
```

```python
# splitting the train into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

# feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train) # fit(calculate mean and std) then transform(perform standardization by centering and scaling)
x_test=sc.transform(x_test) # transform

# create polynomial terms for x
from scipy.stats import rankdata,norm
for col in x.columns:
    x[col+'^2'] = x[col] * x[col]
    x[col+'^3'] = x[col] * x[col] * x[col]
    x[col+'^4'] = x[col] * x[col] * x[col] * x[col]
    # cumulative percentile
    x[col+'_cp'] = rankdata(x[col]).astype('float32')
    # cumulative normal percentile
    x[col+'_cnp'] = norm.cdf(x[col]).astype('float32')

# Recursive Feature Elimination
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
rfe = RFE(logreg,20)
rfe = rfe.fit(x,y)
print(rfe.ranking)
```
